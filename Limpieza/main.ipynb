### Preparacion para un modelo de ML

```python
print(df.head())
```

```python
ml_df = df[['duration_num','release_year','type', 'listed', 'main_country', 'rating']] #Selecciona un subconjunto de columnas relevantes para el análisis de machine learning y las almacena en un nuevo DataFrame llamado ml_df.
```

```python
print(ml_df.head())
```

```python
ml_df = ml_df.dropna()
```

```python
import matplotlib.pyplot as plt

# Contamos los ratings
rating_counts = ml_df['rating'].value_counts()

# Hacemos la grafica de barras
plt.figure()
rating_counts.plot(kind='bar')

plt.xlabel('Rating')
plt.ylabel('Número de títulos')
plt.title('Distribución de Ratings')
plt.xticks(rotation=45)

plt.show()


print(ml_df['rating'].value_counts())

```

```python
top_ratings = ml_df['rating'].value_counts().head(5).index
ml_df = ml_df[ml_df['rating'].isin(top_ratings)]

print(ml_df['rating'].value_counts())
```

```python
# Separamos las características (X) de la variable objetivo (y)
X = ml_df.drop(columns=['rating'])
y = ml_df['rating']


X = pd.get_dummies(X, drop_first=True)
```

```python
# Ocupamos escalar las variables para no tengan diferentes pesos en el modelo de machine learning

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y #Stratify=y asegura que la distribución de clases en el conjunto de entrenamiento y prueba sea similar a la distribución original en y, 
                                                            #lo que es especialmente importante para problemas de clasificación con clases desbalanceadas.
)
```

```python
from sklearn.linear_model import LogisticRegression

# P(Y = k | X) = exp(X * beta_k) / sum(exp(X * beta_j)) for j in all classes

model = LogisticRegression(
    solver='lbfgs', #soporta multiclase nativamente
    max_iter=1000
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure()
sns.heatmap(cm, annot=True, fmt='d')
plt.title("Confusion Matrix")
plt.show()
```

```python
coef_df = pd.DataFrame(
    model.coef_,
    columns=X.columns,
    index=model.classes_
)

coef_df
```
