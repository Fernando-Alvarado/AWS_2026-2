# Limpieza

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce') # Convierte la columna date_added a formato de fecha (datetime). Si algún valor no se puede convertir, 
#                                                                           lo reemplaza por NaN en lugar de generar un error.
print(df['date_added'])
```

```python
df['added_year'] = df['date_added'].dt.year  #Crea una nueva columna con el año en que se añadió el contenido.
df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month
df['added_month_name'] = df['date_added'].dt.month_name()
```

# Limpiemos otra columna

```python
df = df.replace("Not Given", pd.NA)
```

```python
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype('float')
print(df['duration_num'])
```

```python
# unidad (min / Season(s))
df['duration_unit'] = df['duration'].str.extract(r'([A-Za-z]+)')
print(df['duration_unit'])
```
