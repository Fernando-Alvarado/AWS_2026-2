```python
# ============================
# FUNDAMENTOS DE PYTHON – PARTE 7
# Bucles: for y while
# ============================

# 1. Bucle for con range
print("Bucle for básico:")
for i in range(5):
    print(i)


# range(inicio, fin)
print("\nfor con inicio y fin:")
for i in range(2, 6):
    print(i)
```

```python
# range(inicio, fin, paso)
print("\nfor con paso:")
for i in range(0, 10, 2):
    print(i)
```

```python
# 3. Bucle while básico
print("\nBucle while:")
contador = 0

while contador < 5:
    print(contador)
    contador += 1   # SI NO ACTUALIZAS, ES INFINITO
```

```python
# 5. break: salir del bucle
print("\nUso de break:")
# Ciclos 
for i in range(10):
    if i == 5:
        break # Salir del bucle 
    print(i)
```

```python
# 6. continue: saltar una iteración
print("\nUso de continue:")
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# 3. Diccionarios (dict)
persona = {
    "nombre": "Ana",
    "edad": 20,
    "ciudad": "CDMX"
}

print(persona)
print(type(persona))

# Acceso por clave
print(persona["nombre"])
print(persona["edad"])
```

```python
# Modificar / agregar
persona["edad"] = 21
persona["profesion"] = "Ingeniera"
print(persona)
```

```python
# ============================
# FUNDAMENTOS DE PYTHON – FUNCIONES
# ============================

# 1. Función básica
def saludar():
    print("Hola desde una función")
    print("hola")

saludar()   # llamada a la función

# 2. Función con parámetros
def saludar_persona(nombre):
    print("Hola", nombre)

saludar_persona("Ana")
saludar_persona("Carlos")
```

```python
# 3. Función con retorno (return)
def sumar(a, b):
    return a + b

resultado = sumar(3, 5)
print(resultado)
```

```python
# 4. Diferencia CLAVE: print vs return
def mostrar_suma(a, b):
    print(a + b)



def obtener_suma(a, b):
    return a + b

x = mostrar_suma(2, 3)
y = obtener_suma(2, 3)

print(x)   # None
print(y)   # 5
```

```python
# 5. Funciones con valores por defecto
def saludar_con_idioma(nombre, idioma="es"):
    if idioma == "es":
        return f"Hola {nombre}"
    elif idioma == "en":
        return f"Hello {nombre}"
    else:
        return "Idioma no soportado"

print(saludar_con_idioma("Oscar"))
print(saludar_con_idioma("Oscar", "en"))
```

```python
# 6. Variables locales vs globales

mensaje = "Hola global"

def mostrar_mensaje():
    mensaje = "Hola local"
    print(mensaje)

mostrar_mensaje()
print(mensaje)
```

```python
# 7. Función que usa condicionales y bucles
#Declarar funciones
def contar_pares(n):
    pares = []
    for i in range(n + 1):
        if i % 2 == 0:
            pares.append(i)
    return pares

print(contar_pares(10))
```

```python
# Two Sum (indices) — Fácil
#
# Dada una lista nums y un entero target, regresa una tupla (i, j)
# con índices distintos tal que:
# nums[i] + nums[j] == target
#
# Puedes asumir que:
# - Siempre existe una solución
# - Solo hay una solución válida
#
# Firma:
# def two_sum(nums: list[int], target: int) -> tuple[int, int]:
#
# Ejemplos:
# nums = [2, 7, 11, 15], target = 9  -> (0, 1)
# nums = [3, 2, 4], target = 6       -> (1, 2)
```
